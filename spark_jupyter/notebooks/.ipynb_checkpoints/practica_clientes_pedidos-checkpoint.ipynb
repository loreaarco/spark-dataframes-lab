## 1. Lee ambos CSV y muestra su esquema y algunas filas.

```python
# TODO
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

# Creo una sesión de Spark utilizando el patrón builder
spark = SparkSession.builder \
    # Asigno un nombre a la aplicación Spark
    # Este nombre se aparecera en la interfaz web de Spark para identificar el trabajo
    .appName("PracticaClientesPedidos") \
    # Indico la dirección del master de Spark
    # En este caso se conecta a un cluster Spark cuyo nodo maestro está en spark-master
    # usando el puerto 7077 (puerto por defecto de Spark standalone)
    .master("spark://spark-master:7077") \
    # Creo la sesión si no existe o la reutiliza si ya está creada
    .getOrCreate()

# Defino las rutas según el entorno Docker, para que vaya a buscar la informacion que luego saldra en la muestra de informacion
# Estas rutas corresponden al volumen mapeado en el docker-compose.yml:
# ./apps (local) -> /opt/spark-apps (contenedor)
path_clientes = "/opt/spark-apps/datos/clientes.csv"
path_pedidos = "/opt/spark-apps/datos/pedidos.csv"

# 1. Lectura de datos
# Uso 'header=True' porque los archivos tienen nombres de columnas
# Uso 'inferSchema=True' para que Spark detecte automáticamente si son números o fechas
df_clientes = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(path_clientes)

df_pedidos = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(path_pedidos)

# aqui muestro los esquemas
print("--- Esquema de la tabla Clientes ---")
df_clientes.printSchema()

print("\n--- Esquema de la tabla Pedidos ---")
df_pedidos.printSchema()

# aqui muestro el contenido
# muestro las primeras 5 filas de los datos
print("\n--- Muestra de datos: Clientes ---")
df_clientes.show(5)

print("\n--- Muestra de datos: Pedidos ---")
df_pedidos.show(5)
```
